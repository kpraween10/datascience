```python
import math as m
nm=np.arange(0,5,0.1)
fn=2**nm
fn2=m.e**nm
plt.plot(nm,fn,nm,fn2,'r')
plt.grid('True')
plt.xlabel('number')
plt.ylabel('fn=e^n')
plt.title('function graph')
```

```python
time=np.arange(0,10,0.1)
Amp=np.sin(time)
amp=np.cos(time)
plt.plot(time,Amp,'r',time,amp,'g')
plt.grid(True , which='both')
plt.title("sine wave")
plt.ylabel('Amplitude')
plt.xlabel("time(wt)")
plt.legend("SC",ncol=2,loc='upper left')
```
